# Analysis of Test Results

Within this notebook, we provide the code that we used to understand our test results.

## Load dependencies and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

pval_threshold = 0.05
cmp_data = pd.read_csv('data/test_results.csv')

cmp_data['delta_ratio'] = cmp_data['delta']/(cmp_data['TP']+cmp_data['TN']+cmp_data['FP']+cmp_data['FN'])
cmp_data['delta_score_ratio'] = cmp_data['delta_score']/(cmp_data['TP']+cmp_data['TN']+cmp_data['FP']+cmp_data['FN'])
cmp_data.loc[cmp_data['training_as_test'], 'Data'] = 'Training'
cmp_data.loc[~cmp_data['training_as_test'], 'Data'] = 'Test'

cmp_data['Pair'] = cmp_data['algorithm'] + '(' + cmp_data['framework_1']+ ', ' + cmp_data['framework_2'] + ')'
cmp_data['chi2_sig'] = abs(cmp_data['chi2_pval'])<pval_threshold
cmp_data['ks_sig'] = abs(cmp_data['ks_pval'])<pval_threshold

algorithms = cmp_data['algorithm'].unique()
```

## Overview of Results

```python
print('Overall number of algorithms', len(cmp_data['algorithm'].unique()))
print('Overall number of pairs:', len(cmp_data['Pair'].unique()))
print()
print('Overall number of comparison:', len(cmp_data))
print('Comparisons with equal classes:', sum(cmp_data['delta']>0))
print('Comparisons with significant differences:', sum(cmp_data['chi2_sig']))

print()
print('Overall number of comparisons with scores:', sum(cmp_data['delta_score']!=-1))
print('Comparisons with equal scores:', sum(cmp_data['delta_score']>0))
print('Comparisons with significant differences:', sum(cmp_data['ks_sig']))
```

## Results per Algorithm

```python
for algorithm in algorithms:
    alg_data = cmp_data[cmp_data['algorithm']==algorithm]
    print('Results for:', algorithm)
    print()
    print('Overall number of comparison:', len(alg_data))
    print('Comparisons with equal classes on training data:', sum((alg_data['delta']==0) & alg_data['training_as_test']))
    print('Comparisons with equal classes on test     data:', sum((alg_data['delta']==0) & ~alg_data['training_as_test']))
    print('Comparisons with significant differences training data:', sum((abs(alg_data['chi2_pval'])<pval_threshold) & alg_data['training_as_test']))
    print('Comparisons with significant differences test     data:', sum((abs(alg_data['chi2_pval'])<pval_threshold) & ~alg_data['training_as_test']))
    print()
    print('Overall number of comparisons with scores:', sum(alg_data['delta_score']!=-1))
    print('Comparisons with equal scores on training data:', sum((alg_data['delta_score']==0) & alg_data['training_as_test']))
    print('Comparisons with equal scores on test data:', sum((alg_data['delta_score']==0) & ~alg_data['training_as_test']))
    print('Comparisons with significant differences on training data:', sum((abs(alg_data['ks_pval'])<pval_threshold) & alg_data['training_as_test']))
    print('Comparisons with significant differences on test     data:', sum((abs(alg_data['ks_pval'])<pval_threshold) & ~alg_data['training_as_test']))
    print()
    print('delta range')
    print(alg_data.groupby('Pair').min()['delta_ratio'])
    print(alg_data.groupby('Pair').max()['delta_ratio'])
    plt.figure()
    sns.violinplot(data=alg_data, y='Pair', x='delta_ratio', hue='Data', split=True, cut=0)
    plt.title(algorithm)
    plt.show()
    print('delta score range')
    print(alg_data.groupby('Pair').min()['delta_score_ratio'])
    print(alg_data.groupby('Pair').max()['delta_score_ratio'])
    plt.figure()
    sns.violinplot(data=alg_data, y='Pair', x='delta_score_ratio', hue='Data', split=True, cut=0)
    plt.title(algorithm)
    plt.show()
    plt.figure()
    sns.countplot(data=alg_data, hue='Pair', x='chi2_sig')
    plt.show()
    plt.figure()
    sns.countplot(data=alg_data, hue='Pair', x='ks_sig')
    plt.show()
    print()
    print('Full data')
    display(alg_data)
    print()
    print()
    print('-----------------------------------------------------------------------------------')
    print()
    print()
```
