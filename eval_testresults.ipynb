# Analysis of Test Results

Within this notebook, we provide the code that we used to understand our test results.

## Load dependencies and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

pval_threshold = 0.05
cmp_data = pd.read_csv('data/test_results.csv')

cmp_data['delta_ratio'] = cmp_data['delta']/(cmp_data['TP']+cmp_data['TN']+cmp_data['FP']+cmp_data['FN'])
cmp_data['delta_score_ratio'] = cmp_data['delta_score']/(cmp_data['TP']+cmp_data['TN']+cmp_data['FP']+cmp_data['FN'])
cmp_data.loc[cmp_data['training_as_test'], 'Data'] = 'Training'
cmp_data.loc[~cmp_data['training_as_test'], 'Data'] = 'Test'


cmp_data['Pair'] = cmp_data['algorithm'] + '(' + cmp_data['framework_1']+ ', ' + cmp_data['framework_2'] + ')'
cmp_data['Frameworks'] = cmp_data['framework_1'].str.extract('([A-Z_]+)') + ', ' + cmp_data['framework_2'].str.extract('([A-Z_]+)')
cmp_data['chi2_sig'] = abs(cmp_data['chi2_pval'])<pval_threshold
cmp_data['ks_sig'] = abs(cmp_data['ks_pval'])<pval_threshold

algorithms = cmp_data['algorithm'].unique()
```

## Overview of Results

```python
print('Overall number of algorithms', len(cmp_data['algorithm'].unique()))
print('Overall number of pairs:', len(cmp_data['Pair'].unique()))
print()
print('Overall number of comparison:', len(cmp_data))
print('Comparisons with equal classes:', sum(cmp_data['delta']>0))
print('Comparisons with significant differences:', sum(cmp_data['chi2_sig']))

print()
print('Overall number of comparisons with scores:', sum(cmp_data['delta_score']!=-1))
print('Comparisons with equal scores:', sum(cmp_data['delta_score']>0))
print('Comparisons with significant differences:', sum(cmp_data['ks_sig']))
```

## Results per Algorithm

```python
for algorithm in algorithms:
    alg_data = cmp_data[cmp_data['algorithm']==algorithm]
    print('Results for:', algorithm)
    print()
    print('Overall number of comparison:', len(alg_data))
    print('Comparisons with equal classes on training data:', sum((alg_data['delta']==0) & alg_data['training_as_test']))
    print('Comparisons with equal classes on test     data:', sum((alg_data['delta']==0) & ~alg_data['training_as_test']))
    print('Comparisons with significant differences training data:', sum((abs(alg_data['chi2_pval'])<pval_threshold) & alg_data['training_as_test']))
    print('Comparisons with significant differences test     data:', sum((abs(alg_data['chi2_pval'])<pval_threshold) & ~alg_data['training_as_test']))
    print()
    print('Overall number of comparisons with scores:', sum(alg_data['delta_score']!=-1))
    print('Comparisons with equal scores on training data:', sum((alg_data['delta_score']==0) & alg_data['training_as_test']))
    print('Comparisons with equal scores on test data:', sum((alg_data['delta_score']==0) & ~alg_data['training_as_test']))
    print('Comparisons with significant differences on training data:', sum((abs(alg_data['ks_pval'])<pval_threshold) & alg_data['training_as_test']))
    print('Comparisons with significant differences on test     data:', sum((abs(alg_data['ks_pval'])<pval_threshold) & ~alg_data['training_as_test']))
    print()
    print('delta range')
    print(alg_data.groupby('Pair').min()['delta_ratio'])
    print(alg_data.groupby('Pair').max()['delta_ratio'])
    plt.figure()
    sns.violinplot(data=alg_data, y='Pair', x='delta_ratio', hue='Data', split=True, cut=0)
    plt.title(algorithm)
    plt.show()
    print('delta score range')
    print(alg_data.groupby('Pair').min()['delta_score_ratio'])
    print(alg_data.groupby('Pair').max()['delta_score_ratio'])
    plt.figure()
    sns.violinplot(data=alg_data, y='Pair', x='delta_score_ratio', hue='Data', split=True, cut=0)
    plt.title(algorithm)
    plt.show()
    plt.figure()
    sns.countplot(data=alg_data, hue='Pair', x='chi2_sig')
    plt.show()
    plt.figure()
    sns.countplot(data=alg_data, hue='Pair', x='ks_sig')
    plt.show()
    print()
    print('Full data')
    display(alg_data)
    print()
    print()
    print('-----------------------------------------------------------------------------------')
    print()
    print()
```

```python
cmp_data['delta_true'] = cmp_data['delta']>0
cmp_data['delta_score_true'] = cmp_data['delta_score']>0

score_data = cmp_data[cmp_data['delta_score']>=0]

by_framework_ratios = cmp_data.groupby('Frameworks').sum() / cmp_data.groupby('Frameworks').count()

by_framework_ratios_scores = score_data.groupby('Frameworks').sum() / score_data.groupby('Frameworks').count()

fig, axes = plt.subplots(2,2, figsize=(12,8), sharex=False, sharey=False)

counts = cmp_data.groupby('Frameworks').count().iloc[:,1]
labels = [counts.index[i] + ('(' + str(counts.iloc[i]) + ')').rjust(8) for i in range(len(counts))]

ax = sns.barplot(data=by_framework_ratios, y=by_framework_ratios.index, x='delta_true', orient='h', color='gray', ax=axes[0,0])
ax.set_title('Tests with deviations in classes by library')
ax.set_yticklabels(labels)
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

ax = sns.barplot(data=by_framework_ratios, y=by_framework_ratios.index, x='chi2_sig', orient='h', color='gray', ax=axes[0,1])
ax.set_title('Tests with significant $\chi^2$-test results by library')
ax.set_yticklabels([])
ax.set_ylabel(None)
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

counts = score_data.groupby('Frameworks').count().iloc[:,1]
labels = [counts.index[i] + ' (' + str(counts.iloc[i]) + ')' for i in range(len(counts))]

ax = sns.barplot(data=by_framework_ratios_scores, y=by_framework_ratios_scores.index, x='delta_score_true', orient='h', color='gray', ax=axes[1,0])
ax.set_title('Tests deviations in scores by library')
ax.set_yticklabels(labels)
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

ax = sns.barplot(data=by_framework_ratios_scores, y=by_framework_ratios_scores.index, x='ks_sig', orient='h', color='gray', ax=axes[1,1])
ax.set_title('Tests with significant Kolmogorov-Smirnoff-test\nresults by library')
ax.set_yticklabels([])
ax.set_ylabel(None)
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

plt.subplots_adjust(hspace=0.5, wspace=0.1)
plt.savefig('figures/results_by_framework.pdf', bbox_inches='tight')
plt.show()
```

```python
cmp_data['delta_true'] = cmp_data['delta']>0
cmp_data['delta_score_true'] = cmp_data['delta_score']>0

score_data = cmp_data[cmp_data['delta_score']>=0]

by_framework_ratios = cmp_data.groupby('dataset').sum() / cmp_data.groupby('dataset').count()

by_framework_ratios_scores = score_data.groupby('dataset').sum() / score_data.groupby('dataset').count()

fig, axes = plt.subplots(2,2, figsize=(12,8), sharex=False, sharey=False)


labels = ['BC', 'RANDOM', 'UNIFORM', 'WINE']

ax = sns.barplot(data=by_framework_ratios, y=by_framework_ratios.index, x='delta_true', orient='h', color='gray', ax=axes[0,0])
ax.set_title('Tests with deviations in classes by data set')
ax.set_yticklabels(labels)
ax.set_ylabel('Data set')
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

ax = sns.barplot(data=by_framework_ratios, y=by_framework_ratios.index, x='chi2_sig', orient='h', color='gray', ax=axes[0,1])
ax.set_title('Tests with significant $\chi^2$-test results by data set')
ax.set_yticklabels([])
ax.set_ylabel(None)
ax.set_ylabel('Data set')
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

ax = sns.barplot(data=by_framework_ratios_scores, y=by_framework_ratios_scores.index, x='delta_score_true', orient='h', color='gray', ax=axes[1,0])
ax.set_title('Tests with deviations in scores by data set')
ax.set_yticklabels(labels)
ax.set_ylabel('Data set')
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

ax = sns.barplot(data=by_framework_ratios_scores, y=by_framework_ratios_scores.index, x='ks_sig', orient='h', color='gray', ax=axes[1,1])
ax.set_title('Tests with significant Kolmogorov-Smirnoff-test\nresults by data set')
ax.set_yticklabels([])
ax.set_ylabel(None)
ax.set_ylabel('Data set')
ax.set_xlabel('Percentage of Tests')
ax.set_xlim(0,1)

plt.subplots_adjust(hspace=0.5, wspace=0.1)
plt.savefig('figures/results_by_dataset.pdf', bbox_inches='tight')

plt.show()
```

```python
alg_name_map = {
    'GaussianNB': 'GNB',
    'KernelNB': 'KDENB',
    'LRridge': 'RIDGE',
    'LR': 'LR',
    'RFwithDepth': 'RF2',
    'RFwithoutDepth': 'RF1',
    'SVMlinear': 'LSVM',
    'KNN': 'KNN',
    'LRlasso': 'LASSO',
    'MostFrequent': 'DUMMY',
    'SVMpoly': 'PSVM',
    'SVMrbf': 'RBFSVM',
    'MLPmomentum': 'MOMMLP',
    'MLP': 'MLP',
    'MultinomialNB': 'MNB'
}


for algorithm in algorithms:
    alg_data = cmp_data[cmp_data['algorithm']==algorithm]
    if algorithm in ['SVMlinear', 'MostFrequent', 'SVMpoly', 'SVMrbf']:
        height=max(2, len(alg_data['Pair'].unique())/2.5)
        fig, axes = plt.subplots(1,1, figsize=(6,height), sharex=False, sharey=False)
        ax = sns.violinplot(data=alg_data, y='Pair', x='delta_ratio', hue='Data', split=True, cut=0, ax=axes)
        ax.set_xlim(0,1)
        ax.set_xlabel('$\Delta/n$')
        ax.set_yticklabels([label.get_text()[label.get_text().find('(')+1:label.get_text().find(')')] for label in ax.get_yticklabels()])
        plt.title(alg_name_map[algorithm])
    else:
        height=max(2, len(alg_data['Pair'].unique())/2.5)
        fig, axes = plt.subplots(1,2, figsize=(12,height), sharex=False, sharey=False)
        ax = sns.violinplot(data=alg_data, y='Pair', x='delta_ratio', hue='Data', split=True, cut=0, ax=axes[0])
        ax.set_xlim(0,1)
        ax.set_xlabel('$\Delta/n$')
        ax.set_yticklabels([label.get_text()[label.get_text().find('(')+1:label.get_text().find(')')] for label in ax.get_yticklabels()])
        ax = sns.violinplot(data=alg_data, y='Pair', x='delta_score_ratio', hue='Data', split=True, cut=0, ax=axes[1])
        ax.set_xlim(0,1)
        ax.set_xlabel('$\Delta_{score}/n$')
        ax.set_yticklabels([])
        ax.set_ylabel(None)
        ax.get_legend().remove()
        plt.suptitle(alg_name_map[algorithm])
        plt.subplots_adjust(wspace=0.1)
    plt.savefig('figures/results_%s.pdf' % alg_name_map[algorithm], bbox_inches='tight')
    plt.show()
```
